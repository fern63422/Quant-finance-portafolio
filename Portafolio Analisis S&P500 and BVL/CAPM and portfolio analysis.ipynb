CON EL PROPÃ“SITO DE CALCULAR  EL SHARPE RATIO

3ERA MODIFICACIÃ“N - AQUÃ SE CONSIDERARÃA UNA MODIFICACIÃ“N DE UN CAMPO

```python
### ExtracciÃ³n de la Tasa libre de Riesgo - la covencion por los retornos mensuales que usaremos es  la T-Bills de 3 meses de US



start = dt.datetime(2021, 12, 31)
end = dt.datetime(2025, 1, 1)

rf = web.DataReader('DGS3MO', 'fred', start, end)
rf = rf.rename(columns={'DGS3MO': 'rf_tasa_anualizada_sin_porcentaje'})
rf['rf_anual'] = rf['rf_tasa_anualizada_sin_porcentaje'] / 100 ### Esta linea es para agregarle porcentaje
rf['rf_mensual'] = (1 + rf['rf_anual'])**(1/12) - 1 # Llevar la tasa anual a mensual
rf = rf.reset_index()
rf = rf.rename(columns={'DATE': 'fecha'}) ### Cambiar el nombre de la columna
rf.head(10)
```

```python
rf.isnull().sum()
```

```python
### Se encontrÃ³ valores perdidos. Era de esperarse que existieran fechas donde la FED no publique los rendimientos de los T-Bills a 3 meses
rf['rf_tasa_anualizada_sin_porcentaje'] = rf['rf_tasa_anualizada_sin_porcentaje'].ffill() # Aqui tomamos el valor anterior al valor perdido como un "Forward Fill"
rf['rf_anual'] = rf['rf_anual'].ffill()
rf['rf_mensual'] = rf['rf_mensual'].ffill()

rf = rf.reset_index()
rf.head()

```

```python
rf.isnull().sum() ### QuedÃ³ limpio para nuestro horizonte
```

```python
### Agregando la tasa libre de riesgo a los retornos para generar los retornos en exceso - AquÃ­ combinamos tablas
df = pd.merge(df,rf , on = 'fecha', how = 'left')
df
```

REGRESIONES PARA CALCULAR LOS BETAS DE MERCADO

```python
### ElaboraciÃ³ de Betas
import statsmodels.api as sm

market_ticker = 'SPY' ###LA teorÃ­a utiliza al S&P500 como Retorno de mercado

pivot = df.pivot(index='fecha', columns='ticker', values='retorno_en_exceso_mensual') ### CAPM se usa con el retorno en exceso

market = pivot[market_ticker].dropna()

betas = []

for col in pivot.columns:
    if col == market_ticker:
        continue
    y = pivot[col].dropna()
    x = market.loc[y.index]
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    betas.append({'ticker': col, 'beta_market': model.params[market_ticker],
                  'alpha_capm': model.params['const'], 'p_value_beta':model.pvalues[market_ticker] , 'r2_adj': model.rsquared_adj})

df_betas = pd.DataFrame(betas)
df_betas.head()
```

```python
### Uniendo los betas a nuestro summary_final
summary_final = summary_final.merge(df_betas, on='ticker', how='left')
```

```python
import pandas as pd

# BLOQUE AUTOMÃTICO: LIMPIEZA DE SERIES BCRP (formato crudo)


def limpiar_bcrp(path, nombre_variable, categoria): ### Se  generarÃ¡ una funciÃ³n que limpie la base da datos por cada indicador que descarguemos,
    ### Como el BCRP tiene un formato igual en todos los indicadores que descargamos se puede generar una funciÃ³n que nos ayude a automatizar este proceso
    df = pd.read_csv(path, encoding='latin-1')
    df = df.dropna(how='all')
    df.columns = ['fecha_raw', 'valor_raw']
    df = df[~df['fecha_raw'].str.contains('Tasas|IPC|Cambio', na=False)]  # Elimina encabezados textuales

    # ConversiÃ³n de meses abreviados (Ene, Feb, Mar, etc.)
    meses = {
        'Ene': 'Jan', 'Feb': 'Feb', 'Mar': 'Mar', 'Abr': 'Apr', 'May': 'May', 'Jun': 'Jun',
        'Jul': 'Jul', 'Ago': 'Aug', 'Sep': 'Sep', 'Oct': 'Oct', 'Nov': 'Nov', 'Dic': 'Dec'
    }

    def convertir_fecha(x):
        for esp, eng in meses.items():
            x = str(x).replace(esp, eng)
        return pd.to_datetime(x, format='%b%y', errors='coerce')

    df['fecha'] = df['fecha_raw'].apply(convertir_fecha)
    df['valor'] = pd.to_numeric(df['valor_raw'], errors='coerce')
    df = df.dropna(subset=['valor', 'fecha']).sort_values('fecha')

    # Estructura final estandarizada
    df['pais'] = 'PerÃº'
    df['variable'] = nombre_variable
    df['categoria'] = categoria # Add the 'categoria' column here
    return df[['fecha', 'pais', 'variable', 'valor', 'categoria']]

# =========================================================
# LISTA DE SERIES A PROCESAR
# =========================================================

archivos_bcrp = [
    ('https://github.com/Josueed054/programacionparcial/raw/main/tasa_referencia_bcrp_mensuales_20251026_162217.csv', 'Tasa de referencia BCRP', 'Monetaria'),
    ('https://github.com/Josueed054/programacionparcial/raw/main/Inflacion_BCRP_Mensuales_2020_2025-20251026-192309.csv', 'InflaciÃ³n (IPC PerÃº)', 'Monetaria'),
    ('https://github.com/Josueed054/programacionparcial/raw/main/PBI_var_BCRP_Mensuales-20251026-173224.csv', 'PBI', 'Real')
]

# =========================================================
# PROCESAMIENTO AUTOMÃTICO
# =========================================================

peru_data = pd.concat([
    limpiar_bcrp(path, nombre, cat) for path, nombre, cat in archivos_bcrp
], ignore_index=True)

# =========================================================
peru_data
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# ============================================================
# 1ï¸âƒ£  CORRELACIONES - PERÃš
# ============================================================

print("=== CORRELACIONES MACRO-FINANCIERAS: PERÃš ===\n")

# Variables relevantes para PerÃº
vars_pe = [
    'retorno_mensual',
    'Tasa de referencia BCRP',
    'PBI',
    'InflaciÃ³n (IPC PerÃº)'
]

# CÃ¡lculo por ticker
for ticker in acciones_pe_macro['ticker'].unique():
    temp = acciones_pe_macro[acciones_pe_macro['ticker'] == ticker][vars_pe].dropna()
    if len(temp) > 2:  # asegurar suficientes observaciones
        corr = temp.corr().round(3)
        print(f"\n>> {ticker}  (n={len(temp)})")
        print(corr['retorno_mensual'])
    else:
        print(f"\n>> {ticker}: No hay suficientes datos para calcular correlaciÃ³n.")

# --- Mapa de calor general PerÃº
corr_pe_total = acciones_pe_macro[vars_pe].corr().round(2)
plt.figure(figsize=(6, 4))
sns.heatmap(corr_pe_total, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlaciones macro-financieras - PerÃº")
plt.show()


# ============================================================
# 2ï¸âƒ£  CORRELACIONES - ESTADOS UNIDOS
# ============================================================

print("\n\n=== CORRELACIONES MACRO-FINANCIERAS: ESTADOS UNIDOS ===\n")

# Variables relevantes para EE.UU.
vars_us = [
    'retorno_mensual',
    'Tasa polÃ­tica monetaria (Fed Funds)',
    'InflaciÃ³n mensual (CPI % MoM)',
    'Tasa de desempleo'
]

# CÃ¡lculo por ticker
for ticker in acciones_us_macro['ticker'].unique():
    temp = acciones_us_macro[acciones_us_macro['ticker'] == ticker][vars_us].dropna()
    if len(temp) > 2:
        corr = temp.corr().round(3)
        print(f"\n>> {ticker}  (n={len(temp)})")
        print(corr['retorno_mensual'])
    else:
        print(f"\n>> {ticker}: No hay suficientes datos para calcular correlaciÃ³n.")

# --- Mapa de calor general EE.UU.
corr_us_total = acciones_us_macro[vars_us].corr().round(2)
plt.figure(figsize=(6, 4))
sns.heatmap(corr_us_total, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlaciones macro-financieras - EE.UU.")
plt.show()
```
